```python
def parse_endereco(partes):
    if not isinstance(partes, list):
        return pd.Series({'rua': "s/r", 'numero': "s/n", 'bairro': "s/b", 'cidade': "s/c", 'estado': "s/e"})
    
    # Último elemento sempre é "Cidade - Estado"
    cidade_estado = partes[-1].strip().split(' - ')
    cidade = cidade_estado[0].strip()
    estado  = cidade_estado[1].strip() if len(cidade_estado) > 1 else "s/e"

    if len(partes) == 2:
        # [Bairro,  Cidade - Estado]
        bairro = partes[0].strip()
        rua, numero = "s/r", "s/n"

    elif len(partes) == 3:
        # [Rua,  Número - Bairro,  Cidade - Estado]
        rua = partes[0].strip()
        num_bairro = partes[1].strip().split(' - ')
        numero = num_bairro[0].strip()
        bairro = num_bairro[1].strip() if len(num_bairro) > 1 else "s/b"

    else:
        # Casos mais complexos: pega tudo antes do penúltimo como rua
        rua = ', '.join(p.strip() for p in partes[:-2])
        num_bairro = partes[-2].strip().split(' - ')
        numero = num_bairro[0].strip()
        bairro = num_bairro[1].strip() if len(num_bairro) > 1 else "s/b"

    return pd.Series({'rua': rua, 'numero': numero, 'bairro': bairro, 'cidade': cidade, 'estado': estado})


# Aplicar
partes = pd_data_sem_nulos['endereco'].str.split(',')
endereco_dividido = partes.apply(parse_endereco)

pd_data_endereco_divididos = pd.concat([pd_data_sem_nulos, endereco_dividido], axis=1)
```
